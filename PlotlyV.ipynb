```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

colores = ["#008B8B", "#8A2BE2", "#f8766d","#E9967A"]

df=pd.read_csv("dataCGP.csv")
```

```python
dftotales = df[['Producto','Total']]
dftotales2 = dftotales.groupby(by=['Producto']).sum()
dftotales2
```

```python
dftotales22 = dftotales.groupby(by=['Producto']).count()
dftotales22
```

```python
dforigen = df[['Producto','Origen']]
dftotales33 = dforigen.groupby(by=['Origen']).count()
dftotales33
```

# **GRÁFICO DE BARRAS**

```python
x = df.Producto.unique()
y = dftotales.Total

#PLOTLY EXPRESS
fig = px.bar(df, x='Producto', y='Total', color=df.Origen, title="Gráfico de barras")
fig.show()
```

```python
#GRAPH OBJECTS


fig =go.Figure(data=go.Bar(x=x,y=y))
fig.update_layout(title='Gráfico de barras', xaxis_title='Categoria de Producto', yaxis_title='Valores')
fig.show()
```

```python
#GRAPH OBJECTS


fig =go.Figure(data=go.Bar(x=y,y=x, orientation = 'h'))
fig.update_layout(title='Gráfico de barras', xaxis_title='Valores', yaxis_title='Producto')
fig.show()
```

```python
x = df.Producto.unique()
y = dftotales.Total

#PLOTLY EXPRESS
fig = px.bar(df, y='Producto', x='Total', color=df.Origen, orientation='h', title="Gráfico de barras")
fig.show()
```

# **Gráfico de torta o Pie Chart**

```python
labels = df.Producto.unique()
valores = dftotales2.Total

#PLOTLY EXPRESS
fig= px.pie(dftotales2, values=valores, names=labels, title="PIE CHART")
fig.show()
```

```python
fig =go.Figure(data=go.Pie(labels=labels,values=valores))
fig.update_layout(title='Pie chart')
fig.show()
```

# **HISTOGRAMA**

```python
fig=px.histogram(df,x='Cantidad',color="Origen")
fig.show()
```

```python
fig =go.Figure(data=go.Histogram(x=df.Cantidad))
fig.update_layout(title='Histograma', xaxis_title='Valores', yaxis_title='Fecuencia')
fig.show()
```

# **GRÁFICO DE LÍNEAS**

```python
fig= px.line(df, x='Fecha', y='Total',color='Origen')
fig.show()
```

```python
fig =go.Figure(data=go.Scatter(x=df.Fecha,y=df.Total))
fig.update_layout(title='Gráfico de líneas', xaxis_title='Fechas', yaxis_title='Total')
fig.show()
```

# **Gráfico de regresión / dispersión**

```python
fig=px.scatter(df, x='Fecha',y='Total', color='Origen')
fig.show()
```

```python
fig =go.Figure(data=go.Scatter(x=df.Fecha,y=df.Total, mode='markers'))
fig.update_layout(title='Gráfico de dispersión', xaxis_title='Fecha', yaxis_title='Total')
fig.show()
```
