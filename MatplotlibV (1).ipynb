```python
import pandas as pd
import matplotlib.pyplot as plt

colores = ["#008B8B","#1ADFCA","#ED310C"]

df = pd.read_csv("dataCGP.csv")
```

```python
df
```

```python
dftotales = df[['Producto','Total']]
dftotales2= dftotales.groupby(by=['Producto']).sum()
dftotales2
```

```python
dfproductos= df[['Producto','Total']]
dftotales22= dfproductos.groupby(by=['Producto']).count()
dftotales22
```

```python
dforigen= df[['Producto','Origen']]
dftotales33= dforigen.groupby(by=['Origen']).count()
dftotales33
```

# **GRÁFICO DE BARRAS**

```python
labelsP = df['Origen'].unique()

plt.bar(labelsP,dftotales33.Producto,color = colores)
plt.xlabel("Origen")
plt.ylabel("Cantidad")
plt.title("Gráfico de barra")
plt.show()


df.groupby('Origen').size().plot(kind='bar',color = colores)
```

```python
labelsP = df['Origen'].unique()

plt.barh(labelsP,dftotales33.Producto,color = colores)
plt.ylabel("Origen")
plt.xlabel("Cantidad")
plt.title("Gráfico de barra horizontal")
plt.show()
```

# **GRÁFICO DE DISPERSIÓN**

```python
dfcantidad = df[['Cantidad','Total']]

plt.scatter(dfcantidad.Cantidad, dfcantidad.Total)
plt.xlabel("Cantidad")
plt.ylabel("Total")
plt.title("Gráfico de dispersión")
plt.show()
```

# **Histograma**

```python
plt.hist(df.Cantidad, df.Total.max())
plt.xlabel("Cantidad")
plt.ylabel("Frecuencia")
plt.title("Histograma")
plt.show()
```

# **Gráfico de torta**

```python
labels = df.Origen.unique()
sizes = dftotales33.Producto

plt.pie(sizes, labels = labels, autopct='%1.1f%%')
plt.title("Gráfico de torta")
plt.show()

df.groupby('Origen').size().plot(kind='pie', autopct='%1.1f%%')

```

# **Gráfico de líneas**

```python
x = df.Fecha
y= df.Total

plt.plot(x,y)
plt.xlabel('Fecha')
plt.ylabel('Totales')
plt.title('Gráfico de líneas')
plt.xticks(rotation=90)
plt.show()


df.plot(x='Fecha', y='Total', kind='line')
```
